```python
from google.colab import drive
drive.mount('/content/drive')

import os
import zipfile
import pandas as pd

zip_file_path = '/content/drive/MyDrive/subset.zip'
extracted_path = '/content/subset'

with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extracted_path)

directory_path = extracted_path
image_paths = [os.path.join(directory_path, item) for item in os.listdir(directory_path) if item.endswith('.jpg')]
data_frame_images = pd.DataFrame({'image_path': image_paths})
data_frame_images.head()
```

```python
from PIL import Image

def process_image(img_path):
    img = Image.open(img_path)
    img = img.resize((224, 224))
    img_array = np.array(img)
    img_array = img_array / 255.0
    return img_array

data_frame_images['image_array'] = data_frame_images['image_path'].apply(process_image)
data_frame_images['label'] = data_frame_images['image_path'].apply(lambda x: os.path.basename(x).split('_')[0])

data_frame_images.head()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
data_frame_images['encoded_label'] = encoder.fit_transform(data_frame_images['label'])

data_frame_images[['label', 'encoded_label']].drop_duplicates()
```
